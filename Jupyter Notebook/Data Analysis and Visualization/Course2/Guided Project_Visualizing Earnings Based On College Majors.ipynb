```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
recent_grads = pd.read_csv('recent_grads.csv')
first_row = recent_grads.iloc[0]
head = recent_grads.head()
print(head)
print(recent_grads.describe())
```

```python
# drop rows with missing values: DataFrame.dropna()

raw_data_count = recent_grads.size
print(raw_data_count)
recent_grads = recent_grads.dropna(axis=0)
cleaned_data_count = recent_grads.size
print(cleaned_data_count)
```

```python
# DataFrame.plot()
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs. Sample_size', figsize=(5,10))
```
