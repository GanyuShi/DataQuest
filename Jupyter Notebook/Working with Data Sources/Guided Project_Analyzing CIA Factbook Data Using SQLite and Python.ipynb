```python
query = 'SELECT name, population, population_growth, birth_rate, death_rate \
        FROM facts \
        WHERE population <> 0 and population <> 7256490011'
df = pd.read_sql_query(query, conn)
df
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(2,2,1)
ax1.hist(df['population'], bins=20)
ax1.set_title('Population distribution')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(df['population_growth'], bins=20, range=(min(df['population_growth']), max(df['population_growth'])))
ax2.set_title('Population growth distribution')

ax3 = fig.add_subplot(2,2,3)
ax3.hist(df['birth_rate'], bins=20, range=(min(df['birth_rate']), max(df['birth_rate'])))
ax3.set_title('Birth rate distribution')

ax4 = fig.add_subplot(2,2,4)
ax4.hist(df['death_rate'], bins=20, range=(min(df['death_rate']), max(df['death_rate'])))
ax4.set_title('Death rate distribution')

plt.show()
```
